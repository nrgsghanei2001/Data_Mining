```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('transfusion.data')
attrs = ["Recency (months)", "Frequency (times)", "Monetary (c.c. blood)", "Time (months)", "whether he/she donated blood in March 2007"]

for i in range(len(attrs) - 1):
    plt.subplot(2, 2, i + 1)
    plt.boxplot(data[attrs[i]])
    plt.title(attrs[i])

plt.subplots_adjust(left=0.1, bottom=0.3, right=1.4, top=1.2, wspace=0.9, hspace=0.4)
plt.show()
```

```python
import seaborn as sbn

for i in range(len(attrs) - 1):
    for j in range(len(attrs) - 1):
        sbn.lmplot(x=attrs[j], y=attrs[i], data=data, hue="whether he/she donated blood in March 2007")
```
